```python
data1={
  "intents": {
    "greeting": {
      "patterns": ["hi", "hello", "hey", "good morning"],
      "responses": ["Hello! Welcome to the college library.", "Hi! How can I assist you today?"]
    },

    "book_price": {
      "patterns": ["book price", "cost of book", "price of book", "how much is book"],
      "responses": ["The price of most academic books ranges between ₹200 to ₹1000 depending on the subject.", "Please specify the book name to get the exact price."]
    },
    "daily_rent": {
      "patterns": ["rent per day", "daily charge", "book rent", "library rent"],
      "responses": ["The rental charge is ₹5 per day for each book.", "Books are rented at ₹5/day."]
    },
    "rent_calculation": {
      "patterns": ["calculate rent", "total rent", "how much rent", "rent for days"],
      "responses": ["Total rent = number of days × ₹5.", "Please tell me how many days you kept the book."]
    },
    "late_fine": {
      "patterns": ["late fine", "overdue fine", "penalty", "late fee"],
      "responses": ["Late fine is ₹10 per day after the due date.", "Penalty = extra days × ₹10."]
    },
    "book_search": {
      "patterns": ["find book", "search book", "do you have"],
      "responses": ["Please tell me the book name or subject."]
    },
    "issue_book": {
      "patterns": ["issue book", "borrow book"],
      "responses": ["You can issue up to 2 books for 14 days."]
    },
    "return_book": {
      "patterns": ["return book"],
      "responses": ["Please return the book before the due date to avoid fines."]
    },
    "library_timing": {
      "patterns": ["timing", "opening time", "closing time"],
      "responses": ["Library is open from 9 AM to 6 PM."]
    },
    "goodbye": {
      "patterns": ["bye", "exit", "goodbye"],
      "responses": ["Goodbye! Visit again."]
    }
  },
  "books": {
    "programming": {
        "python": "₹500",
        "java": "₹450"
    },
    "science": {
        "physics": "₹400",
        "chemistry": "₹420"
    }
  }
}
```

```python
import random
#intent detection
#first def
def intent_detection(user_input):#fun to get user input
    user_input = user_input.lower()#taking user input in lowercase
    for intent, data in data1["intents"].items():#taking intent and data object in intents
        for pattern in data["patterns"]:
            if pattern in user_input:#takking pattern as userinput
                return intent #showing intent
    return "unknown" #else print
#creating converssiosn fun to-
# Find book price
#second def
def get_price(user_input):
    user_input = user_input.lower()#takking lower case input

    for category in data1["books"]:#cheking
        for book, price in data1["books"][category].items():
            if book.lower() in user_input:
                return f"The price of {book} book is {price}"

    return "Sorry, I couldn't find the book price."#else print

# Rent calculation fun
#third def
def calculate_rent(user_input):#take user input
    words = user_input.split()#spliting

    for word in words:
        if word.isdigit():#cheking given word is digit or not
            days = int(word)#taking int
            rent = days * 5 #mulplaying the no. of days to 5
            return f"Total rent for {days} days is ₹{rent}"

    return "Please tell me number of days."#else print

# Chatbot
#last chatbot prosess
def chatbot():
    while True:
        user_input = input("User: ")

        intent = intent_detection(user_input)#take data into first def ,intents

        if intent == "unknown":#cheking def_first else condition
            print("Chatbot: I don't understand")

        elif intent == "book_price":#cheking def_second condition
            print("Chatbot:", get_price(user_input))#take def_second data

        elif intent == "rent_calculation":
            print("Chatbot:", calculate_rent(user_input))#take def_third data

        else:
            response = random.choice(data1["intents"][intent]["responses"])#random key for randmliy selecting the response
            print("Chatbot:", response)
        if intent == "goodbye" or user_input.lower() in ["bye", "goodbye", "exit"]:
           print("Chatbot: Goodbye!")
           break  #stoping loop after goodbye keyword

# Run chatbot
chatbot()#calling to chatbot
```
